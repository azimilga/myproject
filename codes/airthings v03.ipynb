```python
school_col_name   = 'school_id'
room_col_name     = 'room_id'
date_name         = 'date'
time_name         = 'time'
manuf_col_name    = 'manuf_id'
instr_col_name    = 'instr_id'
serial_col_name   = 'serial_id'
co2_name          = 'co2_a'
temp_data_name    = 'temp_a'
rel_hum_name      = 'rh_a'
voc_name          = 'voc_a'
bright_name        = 'bright_a'
sound_name        = 'sound_a'
pm25_name         = 'pm2.5_a'
pm1_name          = 'pm1_a'
radon_name        = 'rdn_a'
press_name        = 'press_a'

#identify a datetime name : wont be displayed. just coding purpose
datetime_name = 'datetime'

#Folder naming directory
folder_raw         = 'data-raw'
folder_ready       = 'data-ready'
folder_year        = '2024'
folder_project     = 'project'
platform_directory = 'C:/Users/azimilga'

 #convert time format to datetime  : time_zone_converter (df, 'Zone Origin', 'Zone Destination')
def time_zone_converter(df, zone_origin, zone_destination):
  #df['datetime'] = pd.to_datetime((df['date'] + ' ' + df['time']), format='%Y-%m-%d %H:%M:%S')
  #df['DateTime'] = datechange.dt.strftime('%Y-%m-%d %H:%M:%S')

  df['recorded'] = pd.to_datetime(df['recorded'])
  df['date'] = df['recorded'].dt.strftime('%Y-%m-%d')

  # Define the timezones (GMT and Europe/Oslo)
  origin_timezone = pytz.timezone(zone_origin)
  destination_timezone = pytz.timezone(zone_destination)

  # Adjust the datetime to the Oslo timezone and account for daylight saving time
  df['adjusted datetime'] = df['recorded'].dt.tz_localize(origin_timezone).dt.tz_convert(destination_timezone)

  # Extract the adjusted time
  df['time'] = df['adjusted datetime'].dt.strftime('%H:%M')
  df['recorded'] = df['adjusted datetime'].copy() #.dt.strftime('%Y-%m-%d %H:%M')
  df.drop(['adjusted datetime'], axis=1, inplace=True)
  
def datedayhour_creator_noindex(df) :
  datechange = pd.to_datetime((df['date'] + ' ' + df['time']), format='%Y-%m-%d %H:%M:%S')
  df['date'] = datechange.dt.strftime('%Y-%m-%d')
  df['time'] = datechange.dt.strftime('%H:%M')
  df['day'] = datechange.dt.dayofweek
  df['hour'] = datechange.dt.hour
```

```python
folder_path1 = input_file_directory # replace with the folder path
csv_files1 = [f for f in os.listdir(folder_path1) if f.endswith('.csv')and f.startswith(serial_id) ] # get list of csv files

dfs1 = [] # list to store dataframes
for file in csv_files1:
    file_path = os.path.join(folder_path1, file) # get full path of csv file
    df = pd.read_csv(file_path, delimiter = ';').iloc[1:] # read csv file into a dataframe
    dfs1.append(df) # append dataframe to list

# concatenate dataframes in the list into a single dataframe
DF2 = pd.concat(dfs1, ignore_index=False) #.fillna(method='backfill')

#spesific directory and what delimitation it uses
#DF2 = pd.read_csv(input_file_directory + file_name, delimiter = ';')
#DF2 = DF2.drop(['PRESSURE hPa'], axis=1)

#convert time format to datetime
datechange = pd.to_datetime(DF2['recorded'])
DF2['recorded'] = datechange.dt.strftime('%Y-%m-%d %H:%M')

#correct the time different with time correction
DF2['recorded'] = pd.to_datetime(DF2['recorded'])

#create new columns
DF2[time_name] = DF2['recorded'].dt.strftime('%H:%M')
DF2[date_name] = DF2['recorded'].dt.strftime('%Y-%m-%d')
DF2.sort_values('recorded', inplace=True)
```

```python
DF2_first = pd.concat([DF2 [date_name],
                       DF2 [time_name],
                       DF2 ['TEMP °C'],
                       DF2 ['HUMIDITY %'],
                       DF2 ['CO2 ppm'],
                       DF2 ['LIGHT %']],
                      axis=1).fillna(method='ffill').iloc[0::2,:].reset_index(level=0, drop=True)

                      #set how many delays you want to each variables because of they have NaN with QL_First
delay = 1
DF2_sort = pd.concat([DF2['PRESSURE hPa'].iloc[delay:].reset_index(drop=True),
                      DF2['VOC ppb'].iloc[delay:].reset_index(drop=True),
                      DF2['PM1 μg/m3'].iloc[delay:].reset_index(drop=True),
                      DF2['PM2_5 μg/m3'].iloc[delay:].reset_index(drop=True),
                      DF2['SOUND_LEVEL_A dBSPL'].iloc[delay:].reset_index(drop=True)], axis=1)

#we deleted every 2 rows to eliminate NaN at Columns Pressure and VOC. other NaN values in PM1, PM2.5 and Sound Leve, are interpolated.
DF2_sort = DF2_sort.iloc[0::2,:].reset_index(level=0, drop=True)
DF2_sort.interpolate(method='linear', inplace=True)
```

```python
DF2_new = pd.concat([DF2_first,DF2_sort], axis=1).reset_index(level=0, drop=True)

#we move columns named 'Date' and 'Time' to the front side

DF2_new [school_col_name]  = school_name
DF2_new [room_col_name]    = class_dir
DF2_new [manuf_col_name]   = manuf_id
DF2_new [instr_col_name]   = instr_id
DF2_new [serial_col_name]  = serial_id

cols_to_move = [school_col_name, room_col_name, date_name, time_name, manuf_col_name, instr_col_name, serial_col_name  ,'PRESSURE hPa', 'TEMP °C', 'HUMIDITY %', 'CO2 ppm', 'VOC ppb','SOUND_LEVEL_A dBSPL']
DF2_new         = DF2_new[ cols_to_move + [ col for col in DF2_new.columns if col not in cols_to_move ] ]
DF2_new.rename(columns={ 'TEMP °C': temp_data_name ,
                        'HUMIDITY %' : rel_hum_name,
                         'CO2 ppm' : co2_name  ,
                         'VOC ppb' : voc_name  ,
                         'LIGHT %' : bright_name,
                         'PM1 μg/m3':pm1_name,
                         'PM2_5 μg/m3' :pm25_name,
                         'SOUND_LEVEL_A dBSPL':sound_name,
                         'PRESSURE hPa' : press_name  }, inplace=True)
```

```python
#we filterize table with new name of DataFrame, and use specific date and time intervals like above. index has been reset as well.
DF2_filtered = DF2_new.dropna(how='all' , axis=0 ).fillna(method='bfill')
#DF2_filtered = DF2_new[DF2_new['date'].between(spec_date,spec_date)].reset_index(level=0, drop=True)

#put spesific Directory and File Name.
directory_path = export_file_directory+class_dir+'/'
name_time = datetime.strptime((DF2_filtered[time_name].iloc[0]),'%H:%M').strftime('%H%M')
name_date = datetime.strptime((DF2_filtered[date_name].iloc[0]),'%Y-%m-%d').strftime('%Y%m%d')

filename = school_name+'_'+class_dir+ '_'+ instr_id + '_'+serial_id +'_'+name_date +'_'+ name_time + '.csv'

# save dataframe to CSV file in specified directory
DF2_filtered.to_csv(directory_path + filename, index=False)
```
