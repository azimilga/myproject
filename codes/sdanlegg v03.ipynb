```python
DF1 = pd.concat(dfs1, ignore_index=False)
DF1['Timestamp'] = pd.to_datetime(DF1['Timestamp'], format = '%d.%m.%Y %H:%M:%S')
DF1 = DF1.sort_values(by=['Timestamp'], ignore_index=True)
DF1.columns = [ datetime_name ,temp_data_name, temp_set_name, co2_data_name ]
DF1 = DF1.set_index(['datetime']).replace(to_replace = ',', value='.', regex = True).replace(to_replace = '\xa0', value='', regex = True).astype(float).resample('1min').mean()
DF1[temp_set_name] = DF1[temp_set_name] .fillna(method='ffill').fillna(method='bfill')
DF1.loc[:,[co2_data_name, temp_data_name]] = DF1.loc[:,[co2_data_name, temp_data_name]] .interpolate(method='index').fillna(method= 'bfill')
DF1.reset_index(inplace=True)

# Adjust the datetime to the Oslo timezone and account for daylight saving time
#DF1['Name :'] = DF1['Name :'].dt.tz_convert(utc_timezone)
DF1[datetime_name] = DF1[datetime_name].dt.tz_localize(oslo_timezone, nonexistent='shift_forward').dt.tz_convert(utc_timezone)
DF1[datetime_name] = DF1[datetime_name].dt.tz_localize(None)
DF1[date_name]   = DF1[datetime_name].dt.strftime(date_format_show)
DF1[time_name]   = DF1[datetime_name].dt.strftime(time_format_show)
DF1.drop(columns=[datetime_name], inplace = True)

DF1 [school_col_name]  = school_name
DF1 [room_col_name]    = class_dir
DF1 [manuf_col_name]   = manuf_id
DF1 [instr_col_name]   = instr_id
DF1 [serial_col_name]  = serial_id
 #.astype(float).interpolate(method='linear').fillna(method='bfill')
cols_to_move = [school_col_name , room_col_name, date_name, time_name, manuf_col_name, instr_col_name, serial_col_name  ]
DF1         = DF1[ cols_to_move + [ col for col in DF1.columns if col not in cols_to_move ]] #.drop(columns=['Date', 'Time'])
```
